```python
# ============================================================
#  SIMPENAN - OOP: INHERITANCE (Pewarisan)
#  Sistem Informasi Penyakit Tanaman Nasional
#  Konteks: Struktur pengguna pada sistem SIMPENAN
# ============================================================

# ─────────────────────────────────────────────
# KELAS INDUK (Super Class / Parent Class)
# ─────────────────────────────────────────────

class Pengguna:
    """
    Super Class: Pengguna
    Merepresentasikan semua pengguna yang terdaftar di sistem SIMPENAN.
    Atribut dasar yang DIMILIKI SEMUA pengguna disimpan di sini,
    sehingga kelas turunan tidak perlu mendefinisikan ulang.
    """

    def __init__(self, nama, username, email):
        # Atribut umum milik SEMUA pengguna (user biasa, admin, petani)
        self.nama     = nama      # Nama lengkap pengguna
        self.username = username  # Username untuk login ke SIMPENAN
        self.email    = email     # Email terdaftar di sistem

    def info(self):
        # Method dasar untuk menampilkan informasi pengguna
        # Akan di-override (ditimpa) oleh kelas turunan
        print(f"[Pengguna] Nama    : {self.nama}")
        print(f"           Username: {self.username}")
        print(f"           Email   : {self.email}")

    def __str__(self):
        # Representasi string objek saat dicetak dengan print()
        return f"Pengguna({self.username})"
```

```python
# ─────────────────────────────────────────────
# KELAS TURUNAN 1: Admin
# Mewarisi semua atribut dari Pengguna
# ─────────────────────────────────────────────

class Admin(Pengguna):
    """
    Child Class 1: Admin
    Mewarisi atribut nama, username, email dari Pengguna.
    Memiliki atribut TAMBAHAN yang hanya dimiliki Admin.
    Di sistem SIMPENAN, Admin bisa kelola data penyakit & user.
    """

    def __init__(self, nama, username, email, nip_admin):
        # super() memanggil __init__ milik Pengguna (kelas induk)
        # Sehingga kita tidak perlu menulis ulang: self.nama, self.username, self.email
        super().__init__(nama, username, email)

        # Atribut KHUSUS Admin (tidak ada di Pengguna biasa)
        self.nip_admin       = nip_admin  # Nomor Induk Pegawai Admin
        self.penyakit_kelola = []         # Daftar penyakit yang dikelola admin ini

    def info(self):
        # Override method info() dari kelas induk
        # Menampilkan info dasar PLUS info khusus Admin
        super().info()                            # Panggil info() milik Pengguna dulu
        print(f"           NIP     : {self.nip_admin}")  # Tambahkan info Admin
        print(f"           Role    : ADMIN")

    def kelola_penyakit(self, nama_penyakit):
        # Method EKSKLUSIF milik Admin: menambahkan data penyakit baru
        # Setara dengan fungsi tambah_penyakit.php di SIMPENAN
        self.penyakit_kelola.append(nama_penyakit)
        print(f"✅ Admin '{self.username}' menambahkan penyakit: '{nama_penyakit}'")

    def lihat_daftar_penyakit(self):
        # Method untuk melihat semua penyakit yang sudah diinput admin
        if not self.penyakit_kelola:
            print("📋 Belum ada penyakit yang dikelola.")
        else:
            print(f"\n📋 Penyakit dikelola oleh Admin '{self.username}':")
            for i, p in enumerate(self.penyakit_kelola, 1):
                # Tampilkan dengan nomor urut
                print(f"   {i}. {p}")

    def __str__(self):
        # Representasi string khusus untuk Admin
        return f"Admin({self.username}, NIP: {self.nip_admin})"
```

```python
# ─────────────────────────────────────────────
# KELAS TURUNAN 2: Petani
# Mewarisi semua atribut dari Pengguna
# ─────────────────────────────────────────────

class Petani(Pengguna):
    """
    Child Class 2: Petani
    Mewarisi atribut nama, username, email dari Pengguna.
    Memiliki atribut TAMBAHAN yang hanya dimiliki Petani.
    Di sistem SIMPENAN, Petani bisa melihat info penyakit dan melaporkan temuan.
    """

    def __init__(self, nama, username, email, lokasi_sawah, jenis_tanaman):
        # super() memanggil __init__ milik Pengguna (kelas induk)
        # Atribut dasar (nama, username, email) langsung diinisialisasi oleh Pengguna
        super().__init__(nama, username, email)

        # Atribut KHUSUS Petani (tidak ada di kelas Pengguna maupun Admin)
        self.lokasi_sawah  = lokasi_sawah   # Lokasi sawah / kebun petani
        self.jenis_tanaman = jenis_tanaman  # Jenis tanaman yang dibudidayakan
        self.laporan       = []             # Riwayat laporan penyakit yang ditemukan

    def info(self):
        # Override method info() dari kelas induk
        super().info()                                    # Panggil info() milik Pengguna
        print(f"           Sawah   : {self.lokasi_sawah}")
        print(f"           Tanaman : {self.jenis_tanaman}")
        print(f"           Role    : PETANI")

    def laporkan_penyakit(self, nama_penyakit, deskripsi_gejala):
        # Method EKSKLUSIF milik Petani: melaporkan temuan penyakit di sawah
        # Setara dengan fitur laporan/konsultasi di SIMPENAN
        laporan_baru = {
            "penyakit"       : nama_penyakit,    # Nama penyakit yang dilaporkan
            "gejala_temuan"  : deskripsi_gejala, # Gejala yang dilihat petani
            "lokasi"         : self.lokasi_sawah  # Lokasi temuan
        }
        self.laporan.append(laporan_baru)  # Tambahkan ke riwayat laporan
        print(f"📝 Petani '{self.username}' melaporkan: '{nama_penyakit}' di {self.lokasi_sawah}")

    def lihat_riwayat_laporan(self):
        # Method untuk melihat riwayat laporan penyakit petani
        if not self.laporan:
            print("📂 Belum ada laporan yang dibuat.")
        else:
            print(f"\n📂 Riwayat Laporan Petani '{self.username}':")
            for i, lap in enumerate(self.laporan, 1):
                # Tampilkan setiap laporan secara detail
                print(f"   {i}. Penyakit : {lap['penyakit']}")
                print(f"      Gejala   : {lap['gejala_temuan']}")
                print(f"      Lokasi   : {lap['lokasi']}")
                print()

    def __str__(self):
        # Representasi string khusus untuk Petani
        return f"Petani({self.username}, Sawah: {self.lokasi_sawah})"
```

```python
# ════════════════════════════════════════════════════════════
#  DEMO: Struktur Class & Object SIMPENAN
# ════════════════════════════════════════════════════════════

if __name__ == "__main__":

    print("=" * 60)
    print("   SIMPENAN - Demo Inheritance (Pewarisan)")
    print("=" * 60)

    # ── Membuat OBJEK dari kelas Admin ──────────────────────
    # Objek: admin1 → Instance dari class Admin
    admin1 = Admin(
        nama      = "Budi Santoso",
        username  = "budi_admin",
        email     = "budi@simpenan.id",
        nip_admin = "ADM-2024-001"
    )

    # ── Membuat OBJEK dari kelas Petani ─────────────────────
    # Objek: petani1 → Instance dari class Petani
    petani1 = Petani(
        nama          = "Siti Rahayu",
        username      = "siti_petani",
        email         = "siti@gmail.com",
        lokasi_sawah  = "Madiun, Jawa Timur",
        jenis_tanaman = "Padi"
    )

    # Objek kedua dari class Petani
    petani2 = Petani(
        nama          = "Joko Widodo",
        username      = "joko_tani",
        email         = "joko@gmail.com",
        lokasi_sawah  = "Kediri, Jawa Timur",
        jenis_tanaman = "Jagung"
    )

    # ── Menampilkan info masing-masing objek ─────────────────
    print("\n🔷 INFO ADMIN:")
    print("-" * 40)
    admin1.info()       # Memanggil method info() milik Admin (override)

    print("\n🟢 INFO PETANI 1:")
    print("-" * 40)
    petani1.info()      # Memanggil method info() milik Petani (override)

    print("\n🟢 INFO PETANI 2:")
    print("-" * 40)
    petani2.info()      # Memanggil method info() milik Petani (override)
```

```python
    # ── Petani melaporkan temuan penyakit ────────────────────
    print("\n" + "=" * 60)
    print("   Aksi Petani: Laporkan Temuan Penyakit")
    print("=" * 60)
    petani1.laporkan_penyakit(
        "Blast Padi",
        "Daun menunjukkan bercak berbentuk belah ketupat, berwarna coklat keabuan"
    )
    petani1.laporkan_penyakit(
        "Hawar Daun Bakteri",
        "Tepi daun mengering dan menggulung, terdapat eksudat bakteri kuning"
    )
    petani1.lihat_riwayat_laporan()   # Tampilkan riwayat laporan petani
```

```python
    # ── Verifikasi Pewarisan dengan isinstance() ─────────────
    print("=" * 60)
    print("   Verifikasi Struktur Pewarisan")
    print("=" * 60)
    # isinstance() memeriksa apakah objek merupakan turunan dari kelas tertentu
    print(f"admin1  adalah Pengguna? → {isinstance(admin1,  Pengguna)}")   # True, karena Admin mewarisi Pengguna
    print(f"petani1 adalah Pengguna? → {isinstance(petani1, Pengguna)}")   # True, karena Petani mewarisi Pengguna
    print(f"admin1  adalah Admin?    → {isinstance(admin1,  Admin)}")      # True
    print(f"petani1 adalah Petani?   → {isinstance(petani1, Petani)}")     # True
    print(f"admin1  adalah Petani?   → {isinstance(admin1,  Petani)}")     # False! Admin ≠ Petani
```
