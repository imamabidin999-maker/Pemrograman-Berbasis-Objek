```python
# ============================================================
#  SIMPENAN - OOP: POLYMORPHISM (Polimorfisme)
#  Sistem Informasi Penyakit Tanaman Nasional
#  Konteks: Satu method "hitung_risiko()" berperilaku BERBEDA
#           tergantung tipe penyakit (Jamur, Bakteri, Virus)
# ============================================================

# ─────────────────────────────────────────────
# KELAS INDUK (Parent Class): Penyakit
# ─────────────────────────────────────────────

class Penyakit:
    """
    Super Class: Penyakit
    Merepresentasikan data penyakit tanaman di SIMPENAN.
    Setara dengan tabel 'penyakit' di database SIMPENAN:
      → id, nama_penyakit, gejala_utama, ciri_ciri, penanganan
    """

    def __init__(self, nama_penyakit, gejala_utama, penanganan):
        # Atribut dasar: sama persis dengan kolom tabel penyakit di SIMPENAN
        self.nama_penyakit = nama_penyakit  # Nama penyakit tanaman
        self.gejala_utama  = gejala_utama   # Gejala yang tampak pada tanaman
        self.penanganan    = penanganan     # Cara penanganan yang disarankan

    def hitung_risiko(self):
        # Method ini akan di-OVERRIDE oleh masing-masing kelas turunan
        # Inilah inti dari POLIMORFISME: satu nama method, banyak perilaku
        # Di kelas induk, method ini hanya sebagai "template" / kerangka
        raise NotImplementedError(
            f"Method hitung_risiko() harus diimplementasikan oleh kelas turunan!"
        )

    def tampilkan_info(self):
        # Method umum yang menampilkan informasi dasar penyakit
        # Diwarisi oleh semua kelas turunan
        print(f"\n📋 Nama Penyakit : {self.nama_penyakit}")
        print(f"   Gejala Utama  : {self.gejala_utama}")
        print(f"   Penanganan    : {self.penanganan}")

    def __str__(self):
        # Representasi string objek Penyakit
        return f"Penyakit({self.nama_penyakit})"
```

```python
# ─────────────────────────────────────────────
# KELAS TURUNAN 1: PenyakitJamur
# Penyakit yang disebabkan oleh jamur (fungi)
# ─────────────────────────────────────────────

class PenyakitJamur(Penyakit):
    """
    Child Class 1: PenyakitJamur
    Contoh: Blast Padi (Pyricularia oryzae), Busuk Daun Jagung
    Risiko dihitung berdasarkan tingkat kelembaban udara:
    → Semakin lembab, semakin tinggi risiko penyebaran jamur
    """

    def __init__(self, nama_penyakit, gejala_utama, penanganan,
                 jenis_jamur, kelembaban_persen):
        # Panggil __init__ milik kelas induk Penyakit menggunakan super()
        # Tidak perlu tulis ulang: self.nama_penyakit, self.gejala_utama, self.penanganan
        super().__init__(nama_penyakit, gejala_utama, penanganan)

        # Atribut KHUSUS Penyakit Jamur
        self.jenis_jamur       = jenis_jamur        # Nama spesies jamur penyebab
        self.kelembaban_persen = kelembaban_persen  # Kelembaban udara saat ini (%)

    def hitung_risiko(self):
        """
        POLIMORFISME: Override method hitung_risiko() khusus untuk Penyakit Jamur.
        Logika: Jamur tumbuh subur di lingkungan lembab.
        - Kelembaban >= 85% → Risiko SANGAT TINGGI
        - Kelembaban 70-84% → Risiko TINGGI
        - Kelembaban 50-69% → Risiko SEDANG
        - Kelembaban < 50%  → Risiko RENDAH
        """
        print(f"\n🍄 [PenyakitJamur] Menghitung risiko: {self.nama_penyakit}")
        print(f"   Jenis Jamur    : {self.jenis_jamur}")
        print(f"   Kelembaban     : {self.kelembaban_persen}%")  # Tampilkan kondisi lingkungan

        # Logika penilaian risiko berdasarkan kelembaban
        if self.kelembaban_persen >= 85:
            risiko = "🔴 SANGAT TINGGI"   # Kondisi sangat ideal untuk jamur
            skor   = 90                    # Skor risiko 0-100
            saran  = "Segera aplikasikan fungisida sistemik, kurangi irigasi."
        elif self.kelembaban_persen >= 70:
            risiko = "🟠 TINGGI"
            skor   = 70
            saran  = "Pantau tanaman harian, siapkan fungisida preventif."
        elif self.kelembaban_persen >= 50:
            risiko = "🟡 SEDANG"
            skor   = 45
            saran  = "Lakukan monitoring mingguan dan pastikan drainase baik."
        else:
            risiko = "🟢 RENDAH"           # Kelembaban rendah, jamur sulit berkembang
            skor   = 15
            saran  = "Kondisi aman, tetap pantau perubahan cuaca."

        # Tampilkan hasil kalkulasi risiko
        print(f"   Tingkat Risiko : {risiko} (Skor: {skor}/100)")
        print(f"   💡 Saran       : {saran}")
        return skor  # Kembalikan skor numerik untuk keperluan lain
```

```python
# ─────────────────────────────────────────────
# KELAS TURUNAN 2: PenyakitBakteri
# Penyakit yang disebabkan oleh bakteri
# ─────────────────────────────────────────────

class PenyakitBakteri(Penyakit):
    """
    Child Class 2: PenyakitBakteri
    Contoh: Hawar Daun Bakteri (BLB), Busuk Lunak Bakteri
    Risiko dihitung berdasarkan suhu lingkungan:
    → Bakteri patogen tanaman aktif di suhu hangat-panas
    """

    def __init__(self, nama_penyakit, gejala_utama, penanganan,
                 jenis_bakteri, suhu_celsius):
        # Panggil konstruktor kelas induk (Penyakit) dengan super()
        super().__init__(nama_penyakit, gejala_utama, penanganan)

        # Atribut KHUSUS Penyakit Bakteri
        self.jenis_bakteri = jenis_bakteri  # Nama spesies bakteri penyebab
        self.suhu_celsius  = suhu_celsius   # Suhu udara saat ini (°C)

    def hitung_risiko(self):
        """
        POLIMORFISME: Override method hitung_risiko() khusus untuk Penyakit Bakteri.
        Logika: Bakteri patogen paling aktif di suhu 25°C - 35°C.
        - Suhu 28-35°C → Risiko SANGAT TINGGI (zona optimal bakteri)
        - Suhu 22-27°C → Risiko TINGGI
        - Suhu 15-21°C → Risiko SEDANG
        - Suhu < 15°C  → Risiko RENDAH (bakteri tidak aktif)
        """
        print(f"\n🦠 [PenyakitBakteri] Menghitung risiko: {self.nama_penyakit}")
        print(f"   Jenis Bakteri  : {self.jenis_bakteri}")
        print(f"   Suhu Saat Ini  : {self.suhu_celsius}°C")  # Tampilkan kondisi suhu

        # Logika penilaian risiko berdasarkan suhu
        if 28 <= self.suhu_celsius <= 35:
            risiko = "🔴 SANGAT TINGGI"   # Zona suhu optimal untuk bakteri patogen
            skor   = 88
            saran  = "Gunakan bakterisida berbasis tembaga, hindari irigasi overhead."
        elif 22 <= self.suhu_celsius < 28:
            risiko = "🟠 TINGGI"
            skor   = 65
            saran  = "Pastikan tanaman tidak terluka, monitor secara intensif."
        elif 15 <= self.suhu_celsius < 22:
            risiko = "🟡 SEDANG"
            skor   = 40
            saran  = "Lakukan monitoring standar, jaga kebersihan lahan."
        else:
            risiko = "🟢 RENDAH"           # Suhu terlalu dingin untuk aktivitas bakteri
            skor   = 10
            saran  = "Risiko rendah, pertahankan sanitasi lahan yang baik."

        # Tampilkan hasil kalkulasi
        print(f"   Tingkat Risiko : {risiko} (Skor: {skor}/100)")
        print(f"   💡 Saran       : {saran}")
        return skor  # Kembalikan nilai numerik
```

```python
# ─────────────────────────────────────────────
# KELAS TURUNAN 3: PenyakitVirus
# Penyakit yang disebabkan oleh virus tanaman
# ─────────────────────────────────────────────

class PenyakitVirus(Penyakit):
    """
    Child Class 3: PenyakitVirus
    Contoh: Tungro Virus, Mosaik Virus Padi
    Risiko dihitung berdasarkan populasi vektor (serangga pembawa virus):
    → Semakin banyak serangga vektor, semakin cepat penyebaran virus
    """

    def __init__(self, nama_penyakit, gejala_utama, penanganan,
                 nama_vektor, jumlah_vektor_per_meter):
        # Panggil konstruktor kelas induk dengan super()
        super().__init__(nama_penyakit, gejala_utama, penanganan)

        # Atribut KHUSUS Penyakit Virus
        self.nama_vektor             = nama_vektor             # Nama serangga vektor
        self.jumlah_vektor_per_meter = jumlah_vektor_per_meter # Populasi vektor per m²

    def hitung_risiko(self):
        """
        POLIMORFISME: Override method hitung_risiko() khusus untuk Penyakit Virus.
        Logika: Virus menyebar melalui serangga vektor.
        - Vektor >= 20/m² → Risiko SANGAT TINGGI
        - Vektor 10-19/m² → Risiko TINGGI
        - Vektor  3-9/m²  → Risiko SEDANG
        - Vektor  < 3/m²  → Risiko RENDAH
        """
        print(f"\n🦟 [PenyakitVirus] Menghitung risiko: {self.nama_penyakit}")
        print(f"   Vektor Penyebar: {self.nama_vektor}")
        print(f"   Populasi Vektor: {self.jumlah_vektor_per_meter} ekor/m²")  # Data populasi

        # Logika penilaian risiko berdasarkan populasi vektor
        if self.jumlah_vektor_per_meter >= 20:
            risiko = "🔴 SANGAT TINGGI"   # Serangan vektor masif
            skor   = 95
            saran  = "SEGERA lakukan penyemprotan insektisida, isolasi area terinfeksi!"
        elif self.jumlah_vektor_per_meter >= 10:
            risiko = "🟠 TINGGI"
            skor   = 72
            saran  = "Gunakan perangkap serangga dan insektisida selektif."
        elif self.jumlah_vektor_per_meter >= 3:
            risiko = "🟡 SEDANG"
            skor   = 38
            saran  = "Pasang yellow sticky trap, pantau perkembangan populasi vektor."
        else:
            risiko = "🟢 RENDAH"           # Populasi vektor sangat sedikit
            skor   = 12
            saran  = "Populasi vektor terkendali, pertahankan kondisi ini."

        # Tampilkan hasil
        print(f"   Tingkat Risiko : {risiko} (Skor: {skor}/100)")
        print(f"   💡 Saran       : {saran}")
        return skor  # Kembalikan skor
```
